```python
#IMPORT IN Dependencies
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

### Bringing in and looking at the data

```python
#Call in path 
in_path = os.path.join("..", "RawData","police-department-incidents.csv")

#READ IN A CSV FILE
df = pd.read_csv(in_path, encoding="UTF-8")

df.head(2)
```

```python
# Print out the columns
df.columns
```

```python
# Drop some columns
df_new = df[['IncidntNum', 'Category', 'Descript', 'DayOfWeek', 'Date', 'Time','PdDistrict', 'Resolution', 'Address', 'X', 'Y', 'Location', 'PdId','SF Find Neighborhoods', 'Current Police Districts','Current Supervisor Districts', 'Analysis Neighborhoods']]
df_new.head(2)
```

```python
# review all catergories
catg = df_new["Category"].value_counts()
df_catg = catg.rename_axis('unique_Category').reset_index(name='counts')
df_catg
```

```python
labels = df_catg["unique_Category"]
sizes = df_catg["counts"]

plt.pie(sizes, labels=labels, autopct="%1.1f%%", shadow=True, startangle=140)

plt.axis("equal") # forces pie charts to have a square border (no ovals)
plt.show()
```

```python
# review all desc
df_other = df_new.loc[df_new["Category"] == "OTHER OFFENSES", :]
desc = df_other["Descript"].value_counts()
df_desc = desc.rename_axis('unique_Category').reset_index(name='counts')
df_desc.head(10)
```

```python
labels = df_desc["unique_Category"]
sizes = df_desc["counts"]

plt.pie(sizes, labels=labels, autopct="%1.1f%%", shadow=True, startangle=140)

plt.axis("equal") # forces pie charts to have a square border (no ovals)
plt.show()
```

```python
# review all desc
df_details_tbl = df_new.groupby(["Category","Descript"]).size().reset_index(name="Count")
df_details_tbl
```

```python
# Convert the counting details summary to a csv to manipulate in excel
summary_tbl = os.path.join("..","RawData","Summary_details_raw.csv")
df_details_tbl.to_csv(summary_tbl, index=False, header=True)
```

```python
grouped_summary_tbl = os.path.join("..","Data","Summary_details_grouped.csv")
df_custom_tble = pd.read_csv(grouped_summary_tbl, encoding="UTF-8")
df_custom_tble = df_custom_tble [["Descript","Top_25_Catg"]]
df_custom_tble.head()
```

```python
merge_table = pd.merge(df_new, df_custom_tble, on="Descript", how="outer")
merge_table.head()
```

```python
# Convert the merged table to a csv to manipulate in excel
output_file = os.path.join("..","Data","Incident_details.csv")
merge_table.to_csv(output_file, index=False, header=True)
```
